```python
data_all = pd.read_csv('spendVsResults.csv')
```

```python
#NFC divisions
nfc_east = data_all.loc[data_all['Division'] == 'NFC East']
nfc_north = data_all.loc[data_all['Division'] == 'NFC North']
nfc_south = data_all.loc[data_all['Division'] == 'NFC South']
nfc_west = data_all.loc[data_all['Division'] == 'NFC West']
```

```python
#AFC divisions
afc_east = data_all.loc[data_all['Division'] == 'AFC East']
afc_north = data_all.loc[data_all['Division'] == 'AFC North']
afc_south = data_all.loc[data_all['Division'] == 'AFC South']
afc_west = data_all.loc[data_all['Division'] == 'AFC West']
```

```python
#Full NFC standings
standings_nfc = nfc_east.append(nfc_north, ignore_index=False, verify_integrity=False, sort=None) 
standings_nfc2 = standings_nfc.append(nfc_south, ignore_index=False, verify_integrity=False, sort=None)
standings_nfc3 = standings_nfc2.append(nfc_west, ignore_index=False, verify_integrity=False, sort=None)
```

```python
#Full AFC standings
standings_afc = afc_east.append(afc_north, ignore_index=False, verify_integrity=False, sort=None) 
standings_afc2 = standings_afc.append(afc_south, ignore_index=False, verify_integrity=False, sort=None)
standings_afc3 = standings_afc2.append(afc_west, ignore_index=False, verify_integrity=False, sort=None)
```

```python
#Pull only Team, record and W/L %
standings = data_all[['Team','W','L','T','W-L%']]

data_WL = standings[['Team','W','L','T']]

data_wpct = standings[['Team','W-L%']]

data_wpct_sorted = data_wpct.sort_values('W-L%', ascending=False)

#calculate average win percentage
wpct_mean = data_wpct_sorted['W-L%'].mean()

#pare down spend data to team and total spend
data_spend = data_all[['Team','Total Spend']]

#sort spend data
data_spend_sorted = data_spend.sort_values('Total Spend', ascending=False)

#calculate league average salary
salary_mean = data_spend_sorted['Total Spend'].mean()
```

```python
#create bar chart of team spend on salary
data_spend_sorted.plot(kind='bar', x='Team', y='Total Spend', xlabel= 'Team', ylabel= 'Salary Expenses (in hundreds of millions $)', 
                       title = 'Total Salary Spend Per Team (BO)',figsize=(16, 10), color=('b','b','b','b','b','b','y','b','b','b',
                        'b','b','b','b','b','b','b','b','b','b','b','b','b','b','b','b','b','b','y','b','b','b'))

#add league average spend line
plt.axhline(salary_mean, color='k', linestyle='--', linewidth=2, label='Avg')

#add league salary cap line
plt.axhline(182500000, color='r', linestyle='--', linewidth=2, label='Cap')

print('yellow = Made Superbowl')


#1/3 of the teams qualifying for the playoffs spent below the league average (5 of 14 teams)
```

```python
spendVsResults = pd.read_csv('spendVsResults.csv')
spendVsResults
```

```python
sns.set_style("white")
spendVsResults.plot.scatter(x="Offense",y="PF",color=("g","r","r","g","r","r","g","r","g","r","r","g",
                            "r","r","r","g","r","g","g","r","r","g","r","r","r","g","g","r","g","g","g","r"), 
                            xlabel= 'Spend on Offense', ylabel= 'Points Scored For Each Team', title = 'Offense Spend vs Points Scored (GC)',
               figsize=(9, 6),s=100)
sns.regplot(data=spendVsResults,x="Offense",y="PF")
```

The Offense Spend vs Points Scored visual above shows that increased spend on offense is correlated with more points scored by each team. Teams with more points scored were more likely to make playoffs.

```python
sns.set_style("white")
spendVsResults.plot.scatter(x="Defense",y="PA",color=("g","r","r","g","r","r","g","r","g","r","r","g",
                            "r","r","r","g","r","g","g","r","r","g","r","r","r","g","g","r","g","g","g","r"),
                            xlabel= 'Spend on Defense', ylabel= 'Points Scored  Against Each Team', title = 'Defense Spend vs Points Against (GC)',
               figsize=(9, 6),s=100)
sns.regplot(data=spendVsResults,x="Defense",y="PA")
```

The Defense Spend vs Points Scored Against visual above shows that increased spend on defense is correlated with less points scored against each team. Teams with less points scored against were more likely to make playoffs.

```python
sns.set_style("white")
spendVsResults.plot.scatter(x="Total Spend",y="PF",color=("g","r","r","g","r","r","g","r","g","r","r","g",
                            "r","r","r","g","r","g","g","r","r","g","r","r","r","g","g","r","g","g","g","r"),
                           xlabel= 'Total Team Spend', ylabel= 'Points Scored For Each Team', title = 'Total Team Spend vs Points Scored (GC)',
              s=100)
sns.regplot(data=spendVsResults,x="Total Spend",y="PF")
```

The Total Team Spend vs Points Scored visual above shows that increased spend on the entire team is correlated with more points scored by each team. Teams with more points scored were more likely to make playoffs.

```python
sns.set_style("white")
spendVsResults.plot.scatter(x="Total Spend",y="PA",color=("g","r","r","g","r","r","g","r","g","r","r","g",
                            "r","r","r","g","r","g","g","r","r","g","r","r","r","g","g","r","g","g","g","r"),
                            xlabel= 'Total Team Spend', ylabel= 'Points Scored Against Each Team', title = 'Total Team Spend vs Points Against (GC)',
               figsize=(9, 6),s=100)
sns.regplot(data=spendVsResults,x="Total Spend",y="PA")
```

The Total Team Spend vs Points Scored Against visual above shows that increased spend on the entire team is correlated with less points scored against each team. Teams with more less scored againstwere more likely to make playoffs.

Of all the above visuals, the strongest correlation was the Total Spend vs Points Scored, seen by the slope of the best fit line.
