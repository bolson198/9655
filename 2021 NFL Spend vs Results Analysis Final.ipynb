```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
#Analyis of NFL salary cap spending on performance
```

```python
data_all = pd.read_csv('spendVsResults.csv')
```

```python
#NFC divisions
nfc_east = data_all.loc[data_all['Division'] == 'NFC East']
nfc_north = data_all.loc[data_all['Division'] == 'NFC North']
nfc_south = data_all.loc[data_all['Division'] == 'NFC South']
nfc_west = data_all.loc[data_all['Division'] == 'NFC West']
```

```python
#AFC divisions
afc_east = data_all.loc[data_all['Division'] == 'AFC East']
afc_north = data_all.loc[data_all['Division'] == 'AFC North']
afc_south = data_all.loc[data_all['Division'] == 'AFC South']
afc_west = data_all.loc[data_all['Division'] == 'AFC West']
```

```python
#Full NFC standings
standings_nfc = nfc_east.append(nfc_north, ignore_index=False, verify_integrity=False, sort=None) 
standings_nfc2 = standings_nfc.append(nfc_south, ignore_index=False, verify_integrity=False, sort=None)
standings_nfc3 = standings_nfc2.append(nfc_west, ignore_index=False, verify_integrity=False, sort=None)
```

```python
#Full AFC standings
standings_afc = afc_east.append(afc_north, ignore_index=False, verify_integrity=False, sort=None) 
standings_afc2 = standings_afc.append(afc_south, ignore_index=False, verify_integrity=False, sort=None)
standings_afc3 = standings_afc2.append(afc_west, ignore_index=False, verify_integrity=False, sort=None)
```

```python
#Pull only Team, record and W/L %
standings = data_all[['Team','W','L','T','W-L%']]

data_WL = standings[['Team','W','L','T']]

data_wpct = standings[['Team','W-L%']]

data_wpct_sorted = data_wpct.sort_values('W-L%', ascending=False)

#calculate average win percentage
wpct_mean = data_wpct_sorted['W-L%'].mean()

#pare down spend data to team and total spend
data_spend = data_all[['Team','Total Spend']]

#sort spend data
data_spend_sorted = data_spend.sort_values('Total Spend', ascending=False)

#calculate league average salary
salary_mean = data_spend_sorted['Total Spend'].mean()
```

```python
#create bar chart of team spend on salary
data_spend_sorted.plot(kind='bar', x='Team', y='Total Spend', xlabel= 'Team', ylabel= 'Salary Expenses (in hundreds of millions $)', 
                       title = 'Total Salary Spend Per Team (BO)',figsize=(16, 10), color=('b','b','b','b','b','b','y','b','b','b',
                        'b','b','b','b','b','b','b','b','b','b','b','b','b','b','b','b','b','b','y','b','b','b'))

#add league average spend line
plt.axhline(salary_mean, color='k', linestyle='--', linewidth=2, label='Avg')

#add league salary cap line
plt.axhline(182500000, color='r', linestyle='--', linewidth=2, label='Cap')

print('yellow = Made Superbowl')


#1/3 of the teams qualifying for the playoffs spent below the league average (5 of 14 teams)
```
